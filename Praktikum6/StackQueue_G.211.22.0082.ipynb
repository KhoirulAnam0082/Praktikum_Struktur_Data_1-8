```python
# Fungsi untuk Menara Hanoi
def TowerOfHanoi(n, dari, menuju, bantuan):
    if n == 0:
        return
    TowerOfHanoi(n-1, dari, bantuan, menuju)
    print("Pindah disk", n, "dari", dari, "ke", menuju)
    TowerOfHanoi(n-1, bantuan, dari, menuju)

# Driver code untuk Menara Hanoi
N = 3
print("\nLangkah-langkah Menara Hanoi:")
TowerOfHanoi(N, 'A', 'C', 'B')
```

```python
# Fungsi untuk mengurutkan stack
def sortStack(input_stack):
    tmpStack = []
    while len(input_stack) > 0:

      #pop out the first element
        tmp = input_stack[-1]
        input_stack.pop()

      #while temporary stack is not empty
      #and top of stack is samller than temp
        while len(tmpStack) > 0 and tmpStack[-1] < tmp:

            #pop from temporary stack and
            #append it to the input stack
            input_stack.append(tmpStack[-1])
            tmpStack.pop()

        #append temp in temporary of stack
        tmpStack.append(tmp)

    return tmpStack

# Fungsi untuk mengurutkan array menggunakan stack
def sortArrayUsingStacks(arr, n):

  #append array elements to stack
    input_stack = []
    i = 0

    while i < n:
        input_stack.append(arr[i])
        i += 1

  #sort the temporary stack
    tmpStack = sortStack(input_stack)
    i = 0

  #put stack elements in arrp[]
    while i < n:
        arr[i] = tmpStack[-1]
        tmpStack.pop()
        i += 1

    return arr

# Driver code untuk mengurutkan array menggunakan stack
arr = [10, 5, 15, 45]
n = len(arr)

arr = sortArrayUsingStacks(arr, n)
print("\nArray setelah diurutkan:")
i = 0
while i < n:
    print(arr[i], end=" ")
    i += 1
```
